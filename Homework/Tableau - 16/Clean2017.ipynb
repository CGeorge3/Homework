```python
import os
import csv
import pandas as pd
import glob
import datetime as dt
import numpy as np
```

```python
merged17File = os.path.join('resources', 'm2017.csv')
bike17File = pd.read_csv(merged17File)
y2017 = pd.DataFrame(bike17File)
y2017
```

```python
y2017.columns = ["trip_duration", "start_time", "stop_time", "start_station_id", "start_station_name", "start_station_lat", "start_station_lon", "end_station_id", "end_station_name", "end_station_lat", "end_station_lon", "bike_id", "user_type", "birth_year", "gender"]
y2017
```

```python
y2017.fillna(" ", inplace=True)
y2017
```

```python
y2017["start_time"] = pd.to_datetime(y2017.start_time)
y2017["stop_time"] = pd.to_datetime(y2017.stop_time)
y2017["start_time"] = y2017["start_time"].dt.strftime("%m/%d/%Y %H:%M")
y2017["stop_time"] = y2017["stop_time"].dt.strftime("%m/%d/%Y %H:%M")
y2017.head()
```

```python
y2017.to_csv('c2017nh.csv', index=False, header=None)
y2017.to_csv('c2017.csv', index=False, header=True)
```
