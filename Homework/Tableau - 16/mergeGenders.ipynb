```python
import os
import csv
import pandas as pd
import glob
import datetime as dt
import numpy as np
```

```python
csvfiles = glob.glob("Gender/*.csv")
wf = csv.writer(open("Gender/mergedGenders.csv", "w"), delimiter = ",")
```

```python
for files in csvfiles:
#      print(files)
    read = csv.reader(open(files, "r"), delimiter = ",")
    next(read)
    # set up writing of each row into writer variable
    for row in read:
#         print(row)
        wf.writerow(row)
```

```python
mg = os.path.join('Gender', 'mergedGenders.csv')
mGenders = pd.read_csv(mg, encoding = "ISO-8859-1")
mGenders
```

```python
mGenders.columns = ["Year", "Total Bikers", "Male", "Female", "Unknown"]
mGenders
```

```python
g2013 = pd.DataFrame({'Year': 2013, 'Total Bikers': 5614887, 'Male': 64.1, 'Female': 19.73, 'Unknown': 16.17}, index=[0])
# concatenate two dataframe
mGenders2 = pd.concat([g2013,mGenders.iloc[:]]).reset_index(drop=True)
mGenders2
```
